```python
df.drop(['Serial No.'],axis=1,inplace=True)
```

```python
df
```

```python
# jabb hame lower bound and upper bound pata rehta hai to ham us time pe min max scaling karte hai
X=df.drop(['Chance of Admit '],axis=1)
y=df['Chance of Admit ']


```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
```

```python
X_train_scaled=ms.fit_transform(X_train)
X_test_scaled=ms.fit_transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(7,activation='relu',input_dim=7))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
model.fit(X_train_scaled,y_train,epochs=10,validation_split=0.2)
```

```python
y_pred=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
